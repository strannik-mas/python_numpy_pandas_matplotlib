```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import xlabel
```

```python
X = np.linspace(0, 3, 1001, dtype=np.float32)
X
```

```python
Y = X**2*np.exp(-(X**2))
Y
```

```python
Z = np.sin(X)/np.exp(X)
Z
```

```python
plt.plot(X, Y, 'b-')
plt.plot(X, Z, 'r-')
S = np.sin(X)
plt.plot(X, S, 'g-')
plt.show()
```

```python
plt.plot(
    X, Y,
    X, Z
)
plt.show
```

```python
X = np.arange(-5, 6, 0.2)
X
```

```python
Y = 0.5 * X + 4 + np.random.randn(*X.shape)/3
Y
```

```python
plt.plot(X,Y, 'go')
plt.show()
```

```python
#метод наименьших квадратов
N = X.shape[0]
a1 = N*sum(X*Y)
a2 = X.sum() * Y.sum()
a3 = N * sum(X**2)
a4 = sum(X)**2
a = (a1 - a2)/(a3 - a4)
a
```

```python
b = (sum(Y) - a*sum(X))/N
b
```

```python
Y1 = a*X+b
plt.plot(X,Y,'go')
plt.plot(X,Y1,'b--')
plt.show()
```
