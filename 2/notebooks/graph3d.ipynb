```python
import numpy as np
import matplotlib.pyplot as plt
from  mpl_toolkits.mplot3d import Axes3D
```

# Параметрическая кривая

```python
t = np.linspace(-13, 13, 1001, dtype=np.float64)
z = t / 5
r = z**2 + 1
x = r * np.cos(t)
y = r * np.sin(t)
```

```python
fig = plt.figure()
#область отрисовки
# ax = fig.gca(projection = '3d') 
ax = fig.add_subplot(projection = '3d')
ax.plot(x,y,z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot(x,y,z, label='line 1')
ax.plot(x,z,y, label='line 2')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1,projection = '3d')  #1 строчка, 2 столбца, 1- данный рисунок будет первым из них
ax2 = fig.add_subplot(1,2,2,projection = '3d')  #1 строчка, 2 столбца, 1- данный рисунок будет первым из них
ax1.plot(x,y,z)
ax2.plot(x,z,y)
plt.show()
```

# Поверхность

```python
x = np.linspace(-10, 10, 2001)
y = np.linspace(-10, 10, 2001)
x1, y1 = np.meshgrid(x, y)
r2 = x1**2 + y1**2
r = np.sqrt(r2)
z = np.exp(-r2/16) * np.cos(r)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_wireframe(x1,y1, z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_wireframe(x1,y1, z, rcount=20, ccount=20)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x1, y1, z)
plt.show()
```

# Круговая диаграмма

```python
A = [10, 15, 3, 0]
B = [15, 14, 3, 25]
names = ['Bacя', "Петя", "Коля", "Саша"]

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.pie(A, labels=names)
ax2.pie(B, labels=names)
plt.show()
```

# Столбчатая диаграмма

```python
wid = 0.5
idx = np.linspace(0, 10, 4, dtype=np.float32)
plt.bar(idx, A, wid, color='b', label='2010')
plt.bar(idx+wid, B, wid, color='r', label='2015')
plt.xticks(idx+wid/2, names)
plt.legend()
plt.show()
```

```python
wid = 0.5
idx = np.linspace(0, 10, 4, dtype=np.float32)
plt.bar(idx, A, wid, color='b', label='2010')
plt.bar(idx, B, wid, color='r', label='2015', bottom=A)
plt.xticks(idx, names)
plt.legend()
plt.show()
```
